```python
#TODO: APPEND FOLLOWER COUNTS TO THE DATAFRAME
 # find unique user and their follower count store in x_df
unique_users = tweets['id_str'].unique()
print(unique_users) # Found that all tweets are by unique twitter handles, will be a problem
the_df = pd.DataFrame(unique_users,columns=('Users',))
the_df['follower_count']=follower_counts
the_df
```

```python
#TODO: Find Polarity of each tweet by each user, average for each user and append to x_df
unique_user_polarity = []
unique_user_subjectivity = []
for user in unique_users:
  polarities = []
  subjectivities = []
  user_df = annotated_tweets[annotated_tweets['user_id']==user]
  for index, rows in user_df.iterrows():
    content = rows[0]
    polarity = TextBlob(content).polarity
    subjectivity = TextBlob(content).subjectivity
    polarities.append(polarity)
    subjectivities.append(subjectivity)
  try:
    avg_polarity = statistics.mean(polarities)
  except:
    avg_polarity=0
  try:
    avg_subjectivity = statistics.mean(subjectivities)
  except:
    avg_subjectivity=0
  
  unique_user_polarity.append(avg_polarity)
  unique_user_subjectivity.append(avg_subjectivity)

the_df['avg polarity'] = unique_user_polarity
the_df['avg subjectivity'] = unique_user_subjectivity
    
the_df
```

```python
# GET NAME AND USERNAME
usernames = []
names = []
tweet_counts = []
for index,rows in tweets.iterrows():
  username = rows['user']['screen_name']
  usernames.append(username)
  name = rows['user']['name']
  names.append(name)
  tweet_count = rows['user']['statuses_count']
  tweet_counts.append(tweet_count)
the_df['username']=usernames
the_df['names']=names
the_df['tweet_counts']=tweet_counts
```

```python
#GET AVERAGE EXTREMETY
label_df = annotated_tweets.copy()
label_df.drop(0)
# print(label_df)
label_df.groupby(['user_id']).mean()
the_df['category'] = label_df['label']
the_df = the_df.fillna(0)
the_df
```

```python
df = the_df.copy()
df = df.drop('Users',axis=1)
df = df.drop('username',axis=1)
df = df.drop('names',axis=1)
x = df.values #returns a numpy array
min_max_scaler = preprocessing.MinMaxScaler()
x_scaled = min_max_scaler.fit_transform(x)
df = pd.DataFrame(x_scaled)
counts = df.sum(axis = 1, skipna = True)
df['score'] = counts
df['users'] = the_df['Users']
df['names'] = the_df['names']
df['username'] = the_df['username']
df
```
